```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

import warnings
warnings.filterwarnings('ignore')

# data = https://www.kaggle.com/datasets/uciml/iris/  ------> Download your data from this link
```

```python
df = pd.read_csv('Iris.csv')
df.head()
df.drop('Id',axis=1,inplace=True)
```

```python
iris_data=df.copy()
iris_data=iris_data.drop("Species",axis=1)
```

```python
iris_data.info()
```

```python
sns.displot(iris_data['SepalLengthCm'],kind='kde')
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
```

```python
minscaler=MinMaxScaler().set_output(transform="pandas")
minscaler.fit(iris_data)
data=minscaler.transform(iris_data)
data
```

```python
sns.displot(data['SepalLengthCm'],kind='kde')
```

```python
stdscaler=StandardScaler().set_output(transform="pandas")
stdscaler.fit(data)
std_data=stdscaler.transform(data)
std_data
```

```python
sns.displot(std_data['SepalLengthCm'],kind='kde')
```

```python
stdscaler=StandardScaler().set_output(transform="pandas")
stdscaler.fit(iris_data)
std_i_data=stdscaler.transform(iris_data)
std_i_data
sns.displot(std_i_data['SepalLengthCm'],kind='kde')
```
