## Feature Handling

- categorical to numerical

```python
data['class']=data["class"].replace({'hypothyroid':1,'negative':0})
```

```python
data['sex']=data['sex'].replace({'F':0,"M":1})
```

```python
for column in data.columns:
    if  len(data[column].unique())==2:
        data[column] = data[column].replace({'f' : 0, 't' : 1})
        print(column, " \n done with {'f' : 0, 't' : 1}")
    elif len(data[column].unique())==1:
        data[column] = data[column].replace({'f': 0})
        print(column, "\n done with'f' : 0")
    elif len(data[column].unique())==1:
        data[column] = data[column].replace({'t' : 1}) 
        print(column, "\n done with 't' : 1")
                                             
    else:
        print('\n',column,'\n')
```

```python
data
```

## Missing value Handling

- using KNN imputer

```python
from sklearn.impute import KNNImputer
impute=KNNImputer(n_neighbors=3)
impute.fit(data)
data_array=impute.transform(data)
```

```python
data=pd.DataFrame(data=np.round(data_array), columns=data.columns)
```
