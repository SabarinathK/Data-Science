```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Dataset link : https://www.kaggle.com/code/sabarinathk/medical-cost-prediction

```python
#Read
df=pd.read_csv('path of data')
df.sample(10)
```

```python
df.info()
```

```python
print('sex :',df['sex'].unique())
print('smoker :',df['smoker'].unique())
print('region :',df['region'].unique())
```

```python
df['sex']=df['sex'].replace({'male':0,'female':1})
df['smoker']=df['smoker'].replace({'yes':0,'no':1})
df['region']=df['region'].replace({'southwest':0,'southeast':1,'northwest':2,'northeast':3})
```

```python
df
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop(['charges'],axis=1)
y=df['charges']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=123)
```

```python
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
print('train score: ',model.score(X_train,y_train)*100)
print('test score : ',model.score(X_test,y_test)*100)
```

```python
y_pred=pd.DataFrame(model.predict(X_test))
y_pred
```

```python
from sklearn.metrics import mean_squared_error
```

```python
metric=mean_squared_error(y_test,y_pred)
metric
```
